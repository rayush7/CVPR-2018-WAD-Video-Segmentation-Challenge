```python
train_loss = '/home/ayush/Instance_Segmentation/all/Mask_RCNN/WAD_Losses/run_wad20190311T1356-tag-loss.csv'
train_mrcnn_bbox_loss = '/home/ayush/Instance_Segmentation/all/Mask_RCNN/WAD_Losses/run_wad20190311T1356-tag-mrcnn_bbox_loss.csv'
train_mrcnn_class_loss = '/home/ayush/Instance_Segmentation/all/Mask_RCNN/WAD_Losses/run_wad20190311T1356-tag-mrcnn_class_loss.csv'
train_mrcnn_mask_loss = '/home/ayush/Instance_Segmentation/all/Mask_RCNN/WAD_Losses/run_wad20190311T1356-tag-mrcnn_mask_loss.csv'
train_rpn_bbox_loss = '/home/ayush/Instance_Segmentation/all/Mask_RCNN/WAD_Losses/run_wad20190311T1356-tag-rpn_bbox_loss.csv'
train_rpn_class_loss = '/home/ayush/Instance_Segmentation/all/Mask_RCNN/WAD_Losses/run_wad20190311T1356-tag-rpn_class_loss.csv'

val_loss = '/home/ayush/Instance_Segmentation/all/Mask_RCNN/WAD_Losses/run_wad20190311T1356-tag-val_loss.csv'
val_mrcnn_bbox_loss = '/home/ayush/Instance_Segmentation/all/Mask_RCNN/WAD_Losses/run_wad20190311T1356-tag-val_mrcnn_bbox_loss.csv'
val_mrcnn_class_loss = '/home/ayush/Instance_Segmentation/all/Mask_RCNN/WAD_Losses/run_wad20190311T1356-tag-val_mrcnn_class_loss.csv'
val_mrcnn_mask_loss = '/home/ayush/Instance_Segmentation/all/Mask_RCNN/WAD_Losses/run_wad20190311T1356-tag-val_mrcnn_mask_loss.csv'
val_rpn_bbox_loss = '/home/ayush/Instance_Segmentation/all/Mask_RCNN/WAD_Losses/run_wad20190311T1356-tag-val_rpn_bbox_loss.csv'
val_rpn_class_loss = '/home/ayush/Instance_Segmentation/all/Mask_RCNN/WAD_Losses/run_wad20190311T1356-tag-val_rpn_class_loss.csv'
```

```python
#Function to Plot the Loss
def plot_loss(train_loss_data,val_loss_data,tag,plot_path):
    train_loss_df = pd.read_csv(train_loss_data,sep =',')
    val_loss_df = pd.read_csv(val_loss_data,sep=',')
    
    epoch_list = list(train_loss_df['Step'])
    train_loss_value = list(train_loss_df['Value'])
    val_loss_value = list(val_loss_df['Value'])
    
    #val_loss_value = val_loss_value[0:len(train_loss_value)]
    
    plt.plot(epoch_list, train_loss_value, linewidth=2.0,linestyle='-',color='darkcyan',label='Train')
    plt.plot(epoch_list, val_loss_value, linewidth=2.0, linestyle = '-' ,color='darkred',label='Val')
    
    plot_title = tag+'_Plot'
    plt.title( tag +'_Plot '+'for Train And Validation Set')
    plt.xlabel('# of Epochs')
    plt.ylabel(tag)
    plt.legend(loc='best')
    plt.grid(axis='both')
    plt.savefig(os.path.join(plot_path,plot_title))
```

```python
plot_loss(train_loss,val_loss,'MRCNN_Total_Loss','./WAD_Losses/')
```

```python
plot_loss(train_mrcnn_bbox_loss,val_mrcnn_bbox_loss,'MRCNN_BBox_Loss','./WAD_Losses/')
```

```python
plot_loss(train_mrcnn_class_loss,val_mrcnn_class_loss,'MRCNN_Class_Loss','./WAD_Losses/')
```
